```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

def f(x):
    return x * 1**1 + 1 * 2**2 #Escrever equação aqui

def bissection(a,b,tol):
    xl = a
    xr = b
    i = 1
    while (np.abs(xl-xr) >= tol):
        c = (xl+xr)/2.0
        prod = f(xl) * f(c)
        if prod > tol:
            xl = c
            i+= 1
        else:
            if prod < tol:
                xr = c
                i+= 1
    return i,c




answer = bissection(-1000,1000,1e-8)
print ("Bisection Method =" , answer[1])
print("Iterations" = , answer[0])



shortcut = fsolve(f,[-1.5,1.5])
print("Fsolve gives 1stroot at x = " ,shortcut[0])

x = np.linspace(-1000,1000,100)
plt.plot(x,f(x))
plt.grid()
plt.show()
```
